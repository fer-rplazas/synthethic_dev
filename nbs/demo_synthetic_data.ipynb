```python
import sys
sys.path.append('../') # Make code in "tools" folder visible to python interpreter

import numpy as np
from scipy import signal

import matplotlib.pyplot as plt
import seaborn as sns
#sns.set_palette("Set2")
%matplotlib widget

plt.rcParams.update(
    {
        "text.usetex": True,
        "font.family": "serif",
        "font.serif": ["Palatino"],
    }
)

from matplotlib.colors import ListedColormap

plt.rcParams["image.cmap"] = ListedColormap(sns.color_palette(as_cmap=True))


from tools.data_generation import DataGenerator
```

```python
Fs = 2048
```

```python
base_cfg = {
    "betas": None,
    "gammas": None,
    "beta_sharpness": None,
    "phase": None,
    "pac": None,
    "cross-pac": None,
    "phase_shift": None,
    "burst_length": None,
}
```

# Beta Gamma ERD / ERS

```python
beta_gamma_cfg = {
    "betas": [0.7,0.4],
    "gammas": [0.4,0.6],
    "beta_sharpness": None,
    "phase": None,
    "pac": None,
    "cross-pac": None,
    "phase_shift": None,
    "burst_length": None,
}
```

```python
data = DataGenerator(beta_gamma_cfg, 1000, Fs=Fs)
```

```python
idx = np.where(np.diff(data.label))[0][4]

signal0 = data.signal[0,idx-Fs:idx]
signal1 = data.signal[0,idx:idx+Fs]

t = np.linspace(0,1,signal0.size)

fig, ax = plt.subplots(2,1)
ax[0].plot(t,signal0, alpha=0.6, label = 'State 0')
ax[0].plot(t,signal1, alpha=0.6, label = 'State 1')
ax[0].legend()
ax[0].set_title("Time domain signals")
ax[0].set_xlabel('Time [s]')

f,Px_0 = signal.periodogram(signal0, fs=Fs)
f,Px_1 = signal.periodogram(signal1, fs=Fs)

ax[1].plot(f,Px_0, alpha = 0.6, label = 'State 0')
ax[1].plot(f,Px_1, alpha = 0.6, label = 'State 1')
ax[1].set_yscale('log')
ax[1].set_ylim(10e-15)
ax[1].legend()
ax[1].set_title("Frequency Spectra")
ax[1].set_xlabel("Freq [Hz]")

fig.tight_layout()
```

```python
data = DataGenerator(beta_gamma_cfg, 2, Fs=Fs)
```

```python
fig, ax = plt.subplots(2,1)

ax[0].plot(data.t, data.signal[0], label = 'signal')
ax[0].plot(data.t, data.label, label = 'state label')
ax[0].legend()
idx = np.where(np.diff(data.label))[0][4]

ax[0].set_xlim(data.t[idx - 6*Fs], data.t[idx + 6*Fs])
ax[0].set_title("Time domain signal")
ax[0].set_xlabel('Time [s]')


f,Px_1 = signal.periodogram(data.signal[0] * data.label.astype(float), fs=Fs)
f,Px_0 = signal.periodogram(data.signal[0] * np.logical_not(data.label).astype(float), fs=Fs)

ax[1].plot(f,np.log(Px_0),label='State 0', alpha = 0.5)
ax[1].plot(f,np.log(Px_1),label='State 1', alpha = 0.5)
ax[1].legend()
ax[1].set_ylim(-15,7.5)
ax[1].set_xlim(0,200)
ax[1].set_title("Frequency domain spectra")
ax[1].set_xlabel("Freq [Hz]")

fig.tight_layout()
```

# Phase-shift


Changing phase relationships between oscillations of random frequencies in two channels

```python
phase_shift_cfg = {
    "betas": None,
    "gammas": None,
    "beta_sharpness": None,
    "phase": None,
    "pac": None,
    "cross-pac": None,
    "phase_shift": [0.1, 0.6],
    "burst_length": None,
}
```

```python
data = DataGenerator(phase_shift_cfg, 10000, Fs=Fs)
```

```python
idx = np.where(np.diff(data.label))[0][4]

signal0 = data.signal[:,idx-Fs:idx]
signal1 = data.signal[:,idx:idx+Fs]

t = np.linspace(0,1,signal0.shape[1])

fig, ax = plt.subplots(3,1)
ax[0].plot(t,signal0[0], alpha=0.6, label = 'State 0')
ax[0].plot(t,signal0[1], alpha=0.6, label = 'State 0')
ax[0].set_title('Two channels phase-shifted by a given amount (state 0)')

ax[1].plot(t,signal1[0], alpha=0.6, label = 'State 0')
ax[1].plot(t,signal1[1], alpha=0.6, label = 'State 0')
ax[1].set_title('Two channels phase-shifted by a different amount (state 1)')

for ax_ in ax[:2]:
    ax_.set_xlabel('Time [s]')

f,Px_0 = signal.periodogram(signal0[1], fs=Fs)
f,Px_1 = signal.periodogram(signal1[1], fs=Fs)

ax[2].plot(f,Px_0, alpha = 0.6, label = 'State 0')
ax[2].plot(f,Px_1, alpha = 0.6, label = 'State 1')
ax[2].set_title('Power spectrum of phase-shifted signals')
ax[2].set_xlabel("Freq [Hz]")
ax[2].legend()
ax[2].set_yscale('log')
ax[2].set_ylim(10e-15)

fig.tight_layout()
```
